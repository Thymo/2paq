```python
from jsonlines import jsonlines

from clustering_utils import get_clusters, embedd_questions, mask_ner
```

```python
path = '../data/results/triviaqa-dev/baseline_hnsw/incorrect.top5.jsonl'
with jsonlines.open(path) as reader:
#     data = [qa for qa in reader]
    data = [qa['input_qa'] for qa in reader]
```

```python
embedded_question_data = embedd_questions(data,'question')
```

```python
clusters = get_clusters(embedded_question_data, 0.5)
len(clusters)
for cluster in clusters:
    for qa in cluster:
        print(qa['question'])
    print('\n')
```

```python
masked_data = mask_ner(data)
```

```python
embedded_data = embedd_questions(data,'masked')
```

```python
clusters = get_clusters(embedded_data, 0.9)
len(clusters)
for cluster in clusters:
    for qa in cluster:
        # print(qa['question'])
        print(qa['masked'])
    print('\n')
```
